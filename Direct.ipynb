## Merging

```python
# getting the directors

merge1 = pd.read_sql("""
SELECT *
FROM directors as di
INNER JOIN movie_basics as mb
    ON di.movie_id = mb.movie_id

INNER JOIN persons as pe
    ON pe.person_id = di.person_id
    
INNER JOIN movie_akas as ma
    ON di.movie_id = ma.movie_id
""", con)

merge1.head(10)
```

```python
merge1.info()
```

```python
merge1.drop_duplicates(keep='first', inplace=True)
```

```python
merge1.head(10)
```

```python
merge1.duplicated().sum()
```

```python
merge1.to_csv('Data/Directors.csv')
```
